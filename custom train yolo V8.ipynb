# YOLO V8 Training


```python
from ultralytics import YOLO
 
model = YOLO('yolov8m.pt')

# Training.
results = model.train(
   data='custom_data.yaml',
   imgsz=640,
   epochs=250,
   batch=8,
   name='HelmetVioltionDetection-YOLOV8m'
)

print("Training Done. Find trained models in runs.")
```
